## Visualizing Data
  - An interesting way  to display Dataframe
  - With matplotlib

```python
import matplotlib.pyplot as plt
a = [1,2,3,4]
b = [5,6,7,8]
plt.figure(figsize=(5,5))
plt.plot(a,b,color='green')
plt.xlabel('Avg')
plt.ylabel('Sum')
plt.show()
```

```python
marks = [79,82,85,75]
divisions = ['A','B','C','D']
#index = np.arange(len(divisions))
width = 0.30
plt.bar(divisions,marks,color='blue')
```

```python
plt.barh(divisions,marks,color='red')
plt
```

```python
lang = ['Python','Java','C','C++']
rating = [75,70,30,65]
plt.pie(rating,labels = lang,shadow=True,startangle=90,autopct='%1.1f%%')
plt.legend(loc='best')
plt.show()
```
